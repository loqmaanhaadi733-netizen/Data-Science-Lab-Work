```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Marks': [85, 90, 78, 92, 88]
}

df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
plt.bar(df['Student'], df['Marks'])
plt.title('Student Performance')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()
```
